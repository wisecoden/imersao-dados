```python
df.columns = ['ano', 'senioridade', 'contrato', 'cargo',
              'salario', 'moeda', 'salario_em_usd', 'residencia',
              'remoto', 'empresa', 'tamanho_empresa']

print("Colunas do DataFrame traduzidas para português:")
display(df.columns)
```

```python
df['senioridade'].value_counts()
```

```python
df['contrato'].value_counts()
```

```python
df['remoto'].value_counts()
```

```python
df['tamanho_empresa'].value_counts()
```

```python
remoto = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}

df['remoto_traduzido'] = df['remoto'].replace(remoto)

print("Categorias da coluna 'remoto' traduzidas:")
display(df['remoto_traduzido'].value_counts())
```
